# 2. Product Features

## $X \in R^{d\times N}, \overset{\text{feature engineering}}{\Longrightarrow} \Phi(x) \rightarrow R^{d'}$

- ### In kernel method, d' usually larger than d 

## 2.1 Monomial Features (Product Features)

### Given $x \in X$ where most information is contained in the __dth order products__ (monomials) of entries $[x]_j$ of x,

## $[x]_{j_1}[x]_{j_2}...[x]_{j_d} \rightarrow\text{ploynomial classifiers}$

### Where $j_1,...,j_d \in \{1,...,N\}$. 

- ### How to use Monomial Features (Product Features)?

    - First, extracting product features,
    
    - Then, applying learning algorithms to these features.

    - In this way, the process can be treated as preprocess the pattern by mapping into feature space $\mathcal{H}$ of all products of *d* entries.

### 2.1.1 Example - Two Dimensional Pattern

- ## $X = R ^2 \rightarrow\text{collect all monomial feature extractors of degree 2}$

    - ## $\Phi: R \rightarrow \mathcal{H} = R^3$
    
    - ## $([x]_1,[x]_2)\rightarrow([x]_1^2,[x]_2^2,[x]_1[x]_2)$
    
    - ## $<\Phi(x),\Phi(x')>=[x]_1^2[x']_1^2+[x]_2^2[x']_2^2+2[x]_1[x]_2[x']_1[x']_2$

- __Note__: The number of terms in __dth__ order monomial for N-dimensional inputs is calculated based on __[stars and bars](https://en.wikipedia.org/wiki/Stars_and_bars_(combinatorics)__

    - ## $N_{\mathcal{H}}=\binom{d+N-1}{d} =\frac{n!}{k!(n-k)!}$

### 2.1.2 Proposition

### Define $C_d$ to map $x \in R^N$ to the vector $C_d(x)$ whose entries are all possible dth degree ordered products of the entries of x. Then the corresponding kernel computing the dot product of vectors mapped by $C_d$ is:

## $k(x,x') = <C_d(x),C_d(x')> = <x,x'>^d$

__Proof__:

## $<C_d(x),C_d(x')> = \sum_{j_1=1}^N...\sum_{j_d=1}^N[x]_{j_1}...[x]_{j_d}[x']_{j_1}...[x']_{j_d}$

## $= \sum_{j_1=1}^N[x]_{j_1}[x']_{j_1}...\sum_{j_d=1}^N[x]_{j_d}[x']_{j_d} = (\sum_{j=1}^N[x]_{j}[x']_{j})^d = <x,x'>^d$

### Note:

### 1) $C_d \rightarrow$ orderd product; $\Phi_d \rightarrow$ unordered product;

### 2) To make the degree of monomials start from 0 to d:

- ### k(x,x') = $(<x,x'>+1)^d$

### 3) Standardize kernel to make the numeric range within some bounded interval, [-1,1]:

- ### ck(x,x'), where c is scaling factor

## 2.1 Positive Definite Kernels

### 2.1.1 Gram Matrix

### Definition:

### Given a function k : $X^2 \rightarrow K\text{ (where K = C or K = R) and pattern } x_1,...,x_m \in X$, the $m\times m$ matrix K with elements

## $K_{ij} := k(x_i, x_j) \rightarrow\text{is called Gram Matrix (or kernel matrix) of k with respect to }x_1,...x_m$



### 2.1.2 PD Matrix

### Definition:

### A complex $m\times m$ matrix K satisfying

### $\forall c_i \in C, \sum_{i,j}c_i\bar{c_j}K_{ij}\geq 0 \\
\\ \Downarrow\\
1) \text{ is called positive definite,}\\
2) \text {and matrix K is positive definite matrix.}\\
3) \text{and $\sum_{i,j}c_i\bar{c_j}K_{ij}$ is the quadratic form induced by K}$ 

- __Note__: a symmetric matrix is positive definite iif all its eigenvalues are nonnegtive.

### 2.1.3 PD Kernel (Gram Matrix + PD Matrix)

### Definition:

### $\exists X\text{, nonempty set, $\forall x_1,...,x_m \in X$ and $\forall m \in N$}$

### $\text{if function k of $X\times X$, generate a positive definite Gram Matrix}$

### $\Rightarrow\text{k is a kernel}$
